```python
import random as rad

sticks = [0, 0, 0, 0]
hscore = 0
nscore=0
result=['도 (1점)','개 (2점)','걸 (3점)','윷 (4점)','모(5점)']

while True:
    while True:
        for i in range(4):
            sticks[i] = rad.randint(0, 1)
        print(f'흥부 {sticks}:', end=' ')
        if sticks.count(0) == 0: 
            hscore += 5
            print(f'{result[-1]}/(총 {hscore}점) --->') 
            if hscore >= 20: 
                break
            continue
        elif sticks.count(0) == 4:
            hscore += 4
            print(f'{result[-2]}/(총 {hscore}점) --->')
            if hscore >= 20: 
                break
            continue
        else:
            hscore += sticks.count(0)
            # Subtract 1 from the index, not the string itself
            print(f'{result[sticks.count(0)-1]}/(총 {hscore}점) --->')  
            break
    if hscore >= 20: 
        print(f'{"-"*80}\n 흥부 승리 => 흥부:{hscore}, 놀부:{nscore}\n{"-"*80}')
        break
    while True:
        for i in range(4):sticks[i] = rad.randint(0, 1)
        print(f'                                        <--- 놀부 {sticks}:', end=' ')
        if sticks.count(0) == 0: 
            nscore += 5
            print(f'{result[-1]}/(총 {nscore}점)')
            if nscore >= 20: 
                break 
            continue
        elif sticks.count(0) == 4:
            nscore += 4
            print(f'{result[-2]}/(총 {nscore}점)')
            if nscore >= 20: 
                break
            continue
        else:
            nscore += sticks.count(0)
            # Subtract 1 from the index here as well
            print(f'{result[sticks.count(0)-1]}/(총 {nscore}점)')  
            break
    if nscore >= 20: 
        print(f'{"-"*80}\n 놀부 승리 => 흥부:{hscore}, 놀부:{nscore}\n{"-"*80}')
        break
```
