```python
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
j1=pd.read_excel('2011-서울경기.xlsx')
j2=pd.read_excel('2011-강원영서.xlsx')
j3=pd.read_excel('2011-강원영동.xlsx')
j4=pd.read_excel('2011-충남.xlsx')
j5=pd.read_excel('2011-충북.xlsx')
print(j1.head())
```

```python
jungbu1=[]
jungbu1=pd.concat([j1,j2,j3,j4,j5],axis=1,ignore_index=True)
jungbu1=jungbu1.iloc[:,[0,2,5,8,11,14]]
jungbu1
```

```python
jungbu1.columns=['날짜','서울','영서','영동','충남','충북']
jungbu1
```

```python
jungbu1.loc[:,'합계']=jungbu1.loc[:,'서울':'충북'].sum(axis=1)
jungbu1
```

```python
print(jungbu1.isna().sum())
```

```python
import numpy as np

jungbu1_mean=int(np.mean(jungbu1['합계']))
```

```python
plt.figure(figsize=(12,3))
plt.bar(jungbu1['날짜'],jungbu1['합계'],color='#5697BF',edgecolor='gray')
plt.axhline(y=jungbu1_mean,color='r',linestyle='--')
plt.title('2011년 6월~8월 중부지방 일별 강수량')
plt.xlabel('날짜')
plt.ylabel('강수량')
plt.grid(linestyle=':')
plt.show()
```
