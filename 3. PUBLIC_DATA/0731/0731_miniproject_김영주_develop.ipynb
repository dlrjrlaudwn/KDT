※ 엘니뇨(라니냐)의 기준: 엘니뇨·라니냐 감시구역(열대 태평양 Nino 3.4 지역: 5°S ~ 5°N, 170°W ~ 120°W)의 3개월 이동 평균한 해수면온도 편차가 +0.5°C 이상(-0.5°C 이하)으로 5개월 이상 지속될 때 그 첫 달을 엘니뇨(라니냐)의 시작으로 봄

###### rolling(window=3)
- rolling 메서드는 현재 열에 대하여 일정 크기의 창(window)를 이용하여 그 window안의 값을 추가 메서드를 통해 계산
- window : 계산할 창(window)의 크기, 열 기준으로 계산할 경우 행의 수에 해당
- min_periods : 계산할 최소 기간,  window 안의 값의 수가 min_periods의 값보다 작을 경우 NaN 출력

```python
elnino_data['3_month_avg'] = elnino_data['DATA'].rolling(window=3, min_periods=1).mean()

# 3개월 평균이 0.5˚C 이상인 경우를 카운트
count = (elnino_data['3_month_avg'] >= 0.5).sum()

print(f"3개월 평균이 0.5˚C 이상인 경우의 개수: {count}")
```

```python
elnino_data
```

```python
#year, month 합치기
elnino_data['year_month'] = elnino_data['YR'].astype(str) + '-' + elnino_data['MON'].astype(str)
elnino_data['year_month'] = pd.to_datetime(elnino_data['year_month'], format='%Y-%m')
```

```python
elnino_data
```

```python
# 엘니뇨 (+0.5 이상) ,라니냐 (-0.5 이하) 조건 설정
threshold_elnino = 0.5
threshold_laninia = -0.5
sustained_period = 5  # 5개월 이상 지속 조건

# 결과를 저장할 리스트
elnino_starts = []
laninia_starts = []

# 5개월 이상 지속되는 엘니뇨/라니냐 시작 월 추출
for i in range(len(elnino_data) - sustained_period + 1):
    # 5개월씩 나눠서 계산 진행
    avg_period = elnino_data['3_month_avg'].iloc[i:i + sustained_period]

    # 엘니뇨 조건 (+0.5 이상) 만족 시 첫 달을 엘니뇨의 시작으로 봄
    if (avg_period >= threshold_elnino).all():
        elnino_starts.append(elnino_data['year_month'].iloc[i])

    # 라니냐 조건 (-0.5 이하) 만족 시 첫 달을 라니냐의 시작으로 봄
    elif (avg_period <= threshold_laninia).all():
        laninia_starts.append(elnino_data['year_month'].iloc[i])

# 빈도 계산
elnino_freq = len(elnino_starts)
laninia_freq = len(laninia_starts)

print(f"엘니뇨 발생 횟수: {elnino_freq}")
print(f"라니냐 발생 횟수: {laninia_freq}")
```

```python
# 전체 온도 그래프
plt.figure(figsize=(30, 6))
plt.plot(elnino_data['year_month'], elnino_data['3_month_avg'],color='black')

# 엘니뇨 영역 (빨간색)
for start in elnino_starts:
    start_idx = elnino_data[elnino_data['year_month'] == start].index[0]
    plt.fill_between(elnino_data['year_month'][start_idx:start_idx+sustained_period],  #시작 달부터 5개월 색칠하기
                     threshold_elnino,  #기준선 (+0.5˚C)
                     elnino_data['3_month_avg'][start_idx:start_idx+sustained_period],
                     where=(elnino_data['3_month_avg'][start_idx:start_idx+sustained_period] >= threshold_elnino),   #5개월동안 3개월의 평균이 기준을 넘는 지 확인
                     color='red')

# 라니냐 영역 (파란색)
for start in laninia_starts:
    start_idx = elnino_data[elnino_data['year_month'] == start].index[0]
    plt.fill_between(elnino_data['year_month'][start_idx:start_idx+sustained_period],   #시작 달부터 5개월 색칠하기
                     threshold_laninia,     #기준선 (-0.5˚C)
                     elnino_data['3_month_avg'][start_idx:start_idx+sustained_period],
                     where=(elnino_data['3_month_avg'][start_idx:start_idx+sustained_period] <= threshold_laninia),     #5개월동안 3개월의 평균이 기준을 넘는 지 확인
                     color='blue')

plt.axhline(y=threshold_elnino, color='red', linestyle='--', label='엘니뇨')    #엘니뇨 기준선
plt.axhline(y=threshold_laninia, color='blue', linestyle='--', label='라니냐')  #라니냐 기준선
plt.title('엘니뇨·라니뇨 발생 현황')
plt.xlabel('연도')
plt.ylabel('해수면 온도 편차 (°C)')
plt.legend()
plt.tight_layout()
plt.show()
```
