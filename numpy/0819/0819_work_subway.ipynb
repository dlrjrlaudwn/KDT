2. 각 지하철 역별 이용객수의 표준 편차
3. 가장 많은 이용객이 있는 역의 이용객 수
4. 가장 적은 이용객이 있는 역의 이용객 수
5. 이용객 수가 가장 많은 역의 이름
6. 이용객 수가 가장 적은 역의 이름
7. 이용객 수의 분포를 히스토그램으로 시각화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/과제1-데이터예시(seoul.csv).csv')
df
```

```python
#각 지하철 역별 평균 이용객수
df.groupby('역명')['승하차객수'].mean()
```

```python
df.groupby('역명')['승하차객수'].sum()/2
```

```python
#각 지하철 역별 이용객수의 표준 편차
df.groupby('역명')['승하차객수'].std()
```

```python
#가장 많은 이용객이 있는 역의 이용객 수
df.groupby('역명')['승하차객수'].sum().max()
```

```python
#가장 적은 이용객이 있는 역의 이용객 수
df.groupby('역명')['승하차객수'].sum().min()
```

```python
#이용객 수가 가장 많은 역의 이름
df.groupby('역명')['승하차객수'].sum().idxmax()
```

```python
#이용객 수가 가장 적은 역의 이름
df.groupby('역명')['승하차객수'].sum().idxmin()
```

```python
#이용객 수의 분포
plt.hist(df.groupby('역명')['승하차객수'].sum())
plt.show()
```
