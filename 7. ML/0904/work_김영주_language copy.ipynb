```python
import pandas as pd
import os
from collections import Counter
```

```python
train_path='../data/lang/train/'
test_path='../data/lang/test'

train_csv='../data/lang_train.csv'
test_csv='../data/lang_test.csv'
```

```python
#파일 읽기
if not os.path.exists(train_csv):          #파일이 없을 때
    print(f'{train_csv} 파일이 없습니다')
else:
    with open(train_csv,mode='r') as f:    #파일이 있을 때
        all_data=f.read()
        print('[File Data]',all_data)
```

```python
#폴더 안에 있는 파일 가져오기
train_files=os.listdir(train_path)

for file in train_files:
    print(file,train_path+file,os.path.exists(train_path+file))
    with open(train_path+file,mode='r',encoding='UTF8') as f:   
        print(f'[File Data] {f.read()}')
```

```python
#폴더 안에 있는 파일 가져오기
train_files=os.listdir(train_path)

for file in train_files:
    print(file,train_path+file,os.path.exists(train_path+file))
    label=file[:2]
    print(f'label: {label}')

    with open(train_path+file,mode='r',encoding='UTF8') as f:
        if label=='en':
            all_data=f.read()
            
            all_data=all_data.lower()            #대소문자 통일
            all_data=all_data.replace('\n','')   #줄바꿈 제거

            for ch in all_data:                  #알파벳 아닌 거 걸러내기
                if not ord('a')<=ord(ch)<=ord('z'):
                    all_data=all_data.replace(ch,'')
            
            print(f'[File Data] {all_data}')
            print(len(all_data))

            #a to z count
            count=sorted(Counter(all_data).items())
            print(count)
```

```python
#a to z count
data={}
for ch in all_data:
    if ch in data:
        data[ch]=data[ch]+1

    else: data[ch]=1
```
