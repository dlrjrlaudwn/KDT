```python
x=ins['age'].to_list()
y=ins['charges'].to_list()

b1,b2=np.polyfit(x,y,1)

plt.scatter(ins['age'],ins['charges'])
plt.plot(x,np.array(x)*b1+b2,color='red',label='y={:.2f}x+{:.2f}'.format(b1,b2))
# sns.regplot(x='bmi',y='charges',data=ins)
plt.xlabel('age')
plt.ylabel('charges')
plt.legend()
plt.show()
```

```python
est=np.array(x)*b1+b2
r2=r2_score(y,est)
print(r2)
```
