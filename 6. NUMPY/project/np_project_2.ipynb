```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import r2_score
```

```python
ins=pd.read_csv('insurance.csv')
ins
```

```python
ins.describe()
```

```python
plt.scatter(ins['age'],ins['charges'],color='#74AEDB')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
x1=ins[ins['charges']<=15000]['age'].to_list()
y1=ins[ins['charges']<=15000]['charges'].to_list()

b1,b2=np.polyfit(x1,y1,1)

plt.scatter(ins[ins['charges']<=15000]['age'],ins[ins['charges']<=15000]['charges'],color='#74AEDB')
plt.plot(x1,np.array(x1)*b1+b2,color='red',label='y1={:.2f}x1 +{:.2f}'.format(b1,b2))
plt.xlabel('age')
plt.ylabel('charges')
plt.legend()
plt.show()
```

```python
est=np.array(x1)*b1+b2
r2=r2_score(y1,est)
print(r2)
```

```python
colors = {'female': '#EDBAB9', 'male': '#74AEDB'}

for i in ins['sex'].unique():
    subset = ins[ins['sex'] == i]
    plt.scatter(subset['age'], subset['charges'],
                color=colors[i], label=i, alpha=0.7)

plt.xlabel("Age")
plt.ylabel("Charges")
plt.legend(title='sex', loc='upper left')
plt.show()
```

```python
colors = {'yes': '#EDBAB9', 'no': '#74AEDB'}

for smoker_status in ins['smoker'].unique():
    subset = ins[ins['smoker'] == smoker_status]
    plt.scatter(subset['age'], subset['charges'],
                color=colors[smoker_status], label=smoker_status, alpha=0.7)

plt.xlabel("Age")
plt.ylabel("Charges")
plt.legend(title='Smoker', loc='upper left')
plt.show()
```
