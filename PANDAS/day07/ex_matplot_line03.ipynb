[matplotlib 활용 데이터 시각화]

X/Y축 처리

```python
#모듈 로딩
from matplotlib import pyplot as plt
```

```python
#그래프 그리기(기본)
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.show()
```

```python
#그래프 그리기_x축과 y축 범위 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.ylim(bottom=0,top=15)
#plt.xlim(right='2024-06-01',left='2023-12-01')
plt.show()
```

```python
#그래프 그리기_x축과 y축 눈금(tick) 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.xticks(rotation=37)
plt.show()
```

```python
#그래프 그리기_그래프 출력 창(figure) 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.figure(figsize=[10,10])  #창 크기 설정
plt.plot(x_data,y_data,marker='o')
plt.xticks(rotation=29.3)
plt.show()
```

```python
#그래프 그리기_그래프 출력 창(figure) 자동 맞춤
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(x_data,y_data,marker='o')
plt.xticks(rotation=29.3)
plt.tight_layout()  #데이터 내용에 따라 자동 맞춤 설정
plt.show()
```

```python
#그래프 그리기_눈금(ticks) 설정
y_data=[1,3,5,7,9]
x_data=['2024-01-01','2024-02-01','2024-03-01','2024-04-01','2024-05-01']
plt.plot(y_data,marker='o')
plt.ylim(bottom=-1,top=15)      
plt.xlim(left=-1,right=7)                              #x축의 범위
font_config={'fontsize':22,'fontweight':'bold'}
plt.xticks([0,2,4,6],['A','B','C','D'],**font_config)    #x축의 눈금
plt.show()
```
