```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv(r'C:\Users\desktop\EX_PANDAS 06\data\titanic.csv')
titanic
```

```python
titanic.info()
```

```python
data=titanic[['survived','sex','age']]
data
```

```python
data.isna().sum()
```

```python
data['survived'].replace(0,np.nan,inplace=True)
```

```python
data.isna().sum()
```

```python
data2=data.dropna().copy()
data2.info()
```

```python
data2.isna().sum()
```

```python
data2=data2.set_index('sex')
data2
```

```python
df_f=data2.loc['female',:]
df_m=data2.loc['male',:]
```

```python
df_f['age_range']=df_f['age']//10
df_f
```

```python
df_m['age_range']=df_m['age']//10
df_m
```

```python
f_age=df_f.value_counts('age_range')
f_age=sorted(f_age.items())
f_age
```

```python
m_age=df_m.value_counts('age_range')
m_age=sorted(m_age.items())
m_age
```

```python
f_df=pd.DataFrame(f_age,columns=['x','y'])
m_df=pd.DataFrame(m_age,columns=['x','y'])
```

```python
df = pd.merge(f_df, m_df, on='x')
df.set_index('x')
```

```python
plt.plot(df['y_x'],label='female')
plt.plot(df['y_y'],label='male')
plt.legend()
plt.show()
```
