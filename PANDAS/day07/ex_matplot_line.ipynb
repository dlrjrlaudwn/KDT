[matplotlib을 활용한 시각화]

matplotlib.pyplot 모듈 사용

다양한 그래프 관련 함수/클래스 존재

```python
#모듈 로딩
import matplotlib.pyplot as plt
```

```python
#그래프 그리기
plt.plot([1,2,3,4,])   #plot(y축 데이터): 선 그래프
plt.xlabel('data')
plt.ylabel('result')
plt.show()
```

```python
#그래프 그리기_선그래프
#plot(x축 데이터, y축 데이터)
plt.plot([10,20,30],[5,10,15])
plt.show()
```

```python
#그래프 그리기_선그래프
#plot(x축 데이터, y축 데이터,fmf(포멧스트링))
#포멧스트링: 그래프 색상, 두께, 스타일 등 지정
plt.plot([10,20,30],[5,10,15],'r--')
plt.show()
```

```python
#그래프 그리기_선그래프
#plot(x축 데이터, y축 데이터,fmf(포멧스트링))
#포멧스트링: 그래프 색상, 두께, 스타일 등 지정
plt.plot([10,20,30],[5,10,15],'k>--')
plt.show()
```

```python
#그래프 그리기_선그래프
#plot(x축 데이터, y축 데이터,line2d 속성으로 설정)
#line2d 속성: color,marker,linestyle(ls),markersize(ms),linewidt(lw),markeredgecolor(mec).markerwidth(mew) ...
plt.plot([10,20,30],[5,10,15],color='red',marker='o',linestyle='dashed',markersize=10,linewidth=4,mec='blue',mew=5)
plt.show()
```

```python
#그래프 그리기_선그래프
plt.plot([10,20,30],[15,25,35],'bo-')
plt.plot([10,20,30],[18,29,25],'ro--')
plt.plot([10,20,30],[25,25,25],'k^-')
plt.show()
```

```python
#그래프 그리기_선그래프
plt.plot([10,20,30],[15,25,35],'bo-',[10,20,30],[18,29,25],'ro--',[10,20,30],[25,25,25],'k^-')
plt.show()
```

```python
#그래프에 부가적인 기능 추가: 제목, x축/y축 라벨, 범례(그래프의 label 속성 값을 legend에서 사용), ...
plt.plot([10,20,30],[15,25,20],'bo--',label='v')
plt.plot([10,20,30],[5,30,28],'ro--',label='2')
plt.title('[matplotlib line2d]')
plt.xlabel('data')
plt.ylabel('result')
plt.legend()
plt.show()
```

```python
#그래프에 부가적인 기능 추가: 제목, x축/y축 라벨, 범례(그래프의 label 속성 값을 legend에서 사용/loc 속성으로 위치 지정 가능), ...
plt.plot([10,20,30],[15,25,20],'bo--',label='v')
plt.plot([10,20,30],[5,30,28],'ro--',label='2')
plt.title('[matplotlib line2d]')
plt.xlabel('data')
plt.ylabel('result')
plt.legend(loc='center')
plt.show()
```
