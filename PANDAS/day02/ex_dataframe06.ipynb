dataframe에서 행/열 추가

```python
#모듈 로딩
import pandas as pd
```

```python
#데이터 준비
data={'name':['홍길동','마징가','원더우먼'],
      'job':['의적','로봇','영웅'],
      'gender':['남','남','여'],
      'age':[100,200,300],
      'height':[180,300,200]}
```

```python
#df로 데이터 저장
datadf=pd.DataFrame(data)
datadf
```

행 추가: 변수명.loc['추가 할 행 이름']=['추가할 데이터']

       : 변수명.iloc[] XXXXXXXX
       
       : 데이터 개수 == 컬럼 개수 -> 행 추가 가능

         데이터 개수 1개 -> 행 추가 가능

         그 외에는 추가 XXXXXXXXXX

```python
#1개 행 추가(데이터 수가 열 개수와 동일한 경우)
datadf.loc[3]=['베트맨','박쥐','남',40,190]
```

```python
datadf
```

```python
#1개 행 추가: 데이터 수 1개인 경우
datadf.loc[4]=100
datadf
```

```python
#1개 행 추가: 데이터 수 2개인 경우
datadf.loc[5]=['스파이더맨','거미']
datadf
```

열 추가: 변수명['추가할 열 이름'] = ['추가할 데이터']

       : 데이터 개수 == 행 개수 -> 열 추가 가능
       
       : 데이터 개수 1개 -> 열 추가 가능
       
       : 그 외에는 XXXXXXXXXXX
