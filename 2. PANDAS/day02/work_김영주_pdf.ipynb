### 행 인덱스, 열이름 설정

```python
import pandas as pd
```

```python
df=pd.DataFrame([[15,'남','덕영중'],[17,'여','수리중']],index=['준서','예은'],columns=['나이','성별','학교'])
```

```python
print('df ========================\n',df)
```

```python
print('df.index =>',df.index)
print('df.columns ===>',df.columns)
```

```python
df.index=['학생1','학생2']
df.columns=['연령','남녀','소속']
```

```python
print(df)
print('\n')
```

```python
print(df.index)
print('\n')
```

```python
print(df.columns)
```

```python
df2=df.rename(columns={'나이':'연령','성별':'남녀','학교':'소속'})
```

```python
df3=df.rename(index={'준서':'학생1','예은':'학생2'},inplace=True)
```

```python
print(f"df======{df3}\n{df}")
print(f"df2======\n{df2}")
```

### 원소 값 변경

```python
import pandas as pd
```

```python
exam_data={'이름':['서준','우현','인아'],'수학':[90,80,70],'영어':[98,89,95],'음악':[85,95,100],'체육':[100,90,90]}
```

```python
df=pd.DataFrame(exam_data)
```

```python
df.set_index('이름',inplace=True)
print(df)
```

```python
df.iloc[0][3]=80
print(df)
```

```python
df.loc['서준']['체육']=90
print(df)
```

```python
df.loc['서준','체육']=100
print(df)
```

```python
df.loc['서준',['음악','체육']]=50
print(df)
```

```python
df.loc['서준',['음악','체육']]=100,50
print(df)
```
