```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
data=df[['survived','pclass','who']]
data
```

```python
data.info()
```

```python
data.head()
```

```python
#결측치 확인
data.isna().sum()
```

```python
import numpy as np
data['survived'].replace(0,np.nan)
```

```python
#생존자 리스트
data2=data.dropna()
data2
```

```python
#생존자 중 어린이 리스트
data_child=data2[data2['who']=='child']
data_child
```

```python
data_child.info()
```

```python
#생존한 어린이들을 좌석 등급을 기준으로 정렬
data_child.sort_values('pclass')
```

```python
#1등실 어린이
child1=data_child['pclass']==1
data_child[child1]
```

```python
data_child[child1].value_counts()
```

```python
#2등실 어린이
child2=data_child['pclass']==2
data_child[child2]
```

```python
data_child[child2].value_counts()
```

```python
#3등실 어린이
child3=data_child['pclass']==3
data_child[child3]
```

```python
data_child[child3].value_counts()
```

```python
#총 탑승자 중 어린이
child=data['who']=='child'
data[child]
```

```python
#총 생존자
p_survived=data['survived']==1
data[p_survived]
```

총 탑승자 891명 중 어린이는 83명이었고, 그 중 생존한 어린이의 수는 49명입니다.

생존한 어린이 중 1등실 어린이 5명, 2등실 어린이 19명, 3등실 어린이 25명으로

낮은 등실의 어린이일수록 더 많이 살아남은 것을 알 수 있었습니다.

전체 생존자 342명 중 어린이의 수가 83명으로 상당히 낮은 비율을 차지하는 것을 알 수 있었습니다.
