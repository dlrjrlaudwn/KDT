[결측치 처리]

데이터가 없는 부분 체크 후 처리 방법 결정

처리 방법: 제거, 치환

```python
#모듈 로딩
import pandas as pd
```

```python
#데이터 준비
file_name='data.csv'
```

```python
#csv -> df
df=pd.read_csv(file_name)
df
```

```python
#데이터 기본 정보 확인
df.info()
```

```python
#실제 데이터와 컬럼별 타입 확인
df.head()
```

```python
#데이터 전처리_결측치 체크: isnull() , isna()
df.isna()
```

```python
df.isna().sum()
```

```python
#데이터 전처리_결측치 처리(삭제): dropna()  <- 행방향/행 안에 1개의 컬럼이라도 nan이면 삭제
df.dropna()
```

```python
#열방향 삭제
df.dropna(axis=1)
```

```python
#모든 원소가 nan일 때 삭제: how='all'
df.dropna(how='all')
```

```python
df.dropna(axis=1,how='all')
```

```python
#정상데이터가 지정된 개수 이하일 경우 삭제: thresh=(정수)
df.dropna(thresh=4)
```

```python
df.dropna(thresh=3)
```

```python
#특정 컬럼만 nan 여부 검사 진행:subset()
# ex) 나이 분석
df.dropna(subset=['age'])
```

```python
#특정 컬럼만 nan 여부 검사 진행:subset()
# ex) 나이&이름 분석
df.dropna(subset=['age','name'])
```

```python
#데이터 전처리_결측치 처리(대체/치환): fillna()  <- 원하는 값, 평균값, 최빈값, 이전 값,이후 값 등 사용 가능
df.fillna(0)
```

```python
df.fillna({'name':'unknown','age':0})
```

```python
#이전 컬럼 값으로 채우기: method='ffill'
df.fillna(method='ffill')
```

```python
df.ffill()
```

```python
#아래 컬럼 값으로 채우기: bfill  <- 아래 컬럼이 없으면 채울 수 없음
df.bfill()
```

```python
#나이 결측치는 나이 평균값으로 채우기
df.fillna({'age':df['age'].mean()})
```

```python
#이름 결측치는 이름 최빈값으로 채우기
df.fillna({'name':df['name'].mode()})
```
