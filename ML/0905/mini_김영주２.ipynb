데이터 전처리

```python
import pandas as pd
```

```python
game=pd.read_csv('/content/vgsales.csv', usecols=[2,4,5,6,7,8,10])
game
```

```python
game.isna().sum()
```

```python
game=game.dropna()
game
```

```python
platform_counts = game['Platform'].value_counts()
popular_platforms = platform_counts[platform_counts > 200].index
filtered_game = game[game['Platform'].isin(popular_platforms)]
filtered_game
```

```python
filtered_game.value_counts('Platform')
```

```python
for i in filtered_game.columns:
    print(i,filtered_game[i].unique())
```

```python
filtered_game[filtered_game['Publisher'] == 'Unknown']['Publisher'].value_counts()
```

```python
filtered_game.drop(filtered_game[filtered_game['Publisher'] == 'Unknown'].index, inplace=True)
filtered_game
```

```python
publisher_count = filtered_game['Publisher'].value_counts()
popular_publisher = publisher_count[publisher_count > 15].index
game_df = filtered_game[filtered_game['Publisher'].isin(popular_publisher)]
game_df
```

```python
game_df.value_counts('Publisher')
```

```python
game_df.value_counts('Platform')
```

feature/target 분리

```python
feature_df=game_df.drop('Platform',axis=1)
target_sr=game_df['Platform']
```

인코딩
- feature: 원 핫 인코딩
- target: 라벨 인코딩

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
label=LabelEncoder()
label.fit(target_sr)
```

```python
target_label=label.transform(target_sr)
target_label
```

```python
feature_ohe=pd.get_dummies(feature_df)
feature_ohe
```

학습용 데이터 셋 준비

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler=StandardScaler()
scaler.fit(feature_ohe)
scaler.transform(feature_ohe)
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature_ohe,target_label,test_size=0.2,random_state=0,stratify=target_label)
```

학습 진행: 로지스틱 회귀

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
print(f'classes_: {lr.classes_}, {len(lr.classes_)}개')
print(f'n_features_in_: {lr.n_features_in_}')
print(f'n_iter_: {lr.n_iter_}')
print(f'coef_: {lr.coef_}, {len(lr.coef_[0])}개')
print(f'intercept_: {lr.intercept_}')
```

```python
train_score=lr.score(x_train,y_train)
test_score=lr.score(x_test,y_test)

print(f'train_score: {train_score}, test_score: {test_score}')
```

```python
lr_pred=lr.predict(x_test)
lr_pred_proba=lr.predict_proba(x_test)
```

```python
from sklearn.metrics import accuracy_score,roc_auc_score
```

```python
print('accuracy:{0:.3f},roc_auc:{1:.3f}'.format(accuracy_score(y_test,lr_pred),
                                                roc_auc_score(y_test,lr_pred_proba,multi_class='ovr')))
```

```python
solvers=['lbfgs','liblinear','newton-cg','sag','saga']

for solver in solvers:
    lr_clf=LogisticRegression(solver=solver,max_iter=600)
    lr_clf.fit(x_train,y_train)
    lr_pred=lr_clf.predict(x_test)
    lr_pred_proba=lr_clf.predict_proba(x_test)

    print('solver: {0}, accuracy: {1:.3f}, roc_auc: {2:.3f}'.format(solver,
                                                                    accuracy_score(y_test,lr_pred),
                                                                    roc_auc_score(y_test,lr_pred_proba,multi_class='ovr')))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test, lr_pred)
```

```python
class_report=classification_report(y_test, lr_pred)
print(class_report)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'penalty':['l2','l1'],
        'C':[0.01,0.1,1,5,10]}

lr_clf=LogisticRegression()

grid_clf=GridSearchCV(lr_clf,param_grid=params,scoring='accuracy',cv=3)
grid_clf.fit(x_train,y_train)
print('최적 하이퍼 파라미터: {0}, 최적 평균 정확도: {1:.3f}'.format(grid_clf.best_params_,
                                                                grid_clf.best_score_))
```
