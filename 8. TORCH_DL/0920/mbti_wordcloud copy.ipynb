```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from wordcloud import WordCloud
```

```python
mbti_df=pd.read_csv('../data/MBTI.csv')
mbti_df
```

```python
mbti_df['ns_type']=mbti_df['type'].apply(lambda x: 1 if 'N' in str(x) else 0)  # 'I'면 1, 'E'면 0
mbti_df['ns_type'].value_counts()
```

```python
mbti_df['posts']=mbti_df['posts'].str.replace(' ',',')
mbti_df
```

```python
n_df=mbti_df[mbti_df['ns_type']==1]
n_df
```

```python
n_word_list=n_df['posts'].to_list()
```

```python
n_dict={}
for word_list in n_word_list:
  for word in word_list.split(','):
    if word in n_dict:
      n_dict[word]+=1
    else:
      n_dict[word]=1
```

```python
n_word_count=Counter(n_dict)
n_word_count.most_common(100)
```

```python
len(n_dict)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(WordCloud(background_color='white',colormap='tab20').generate_from_frequencies(n_dict))
plt.axis('off')
plt.show()
```

```python
s_df=mbti_df[mbti_df['ns_type']==0]

s_word_list=s_df['posts'].to_list()

s_dict={}
for word_list in s_word_list:
  for word in word_list.split(','):
    if word in s_dict:
      s_dict[word]+=1
    else:
      s_dict[word]=1

print(len(s_dict))
```

```python
s_word_count=Counter(s_dict)
s_word_count.most_common(100)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(WordCloud(background_color='white',colormap='tab20').generate_from_frequencies(s_dict))
plt.axis('off')
plt.show()
```

```python
only_n=set(n_dict.keys())-set(s_dict.keys())
only_n
```

```python
len(only_n)
```

```python
only_n_count={}
for word in only_n:
  only_n_count[word]=n_dict[word]

only_n_count
```

```python
only_n_count=Counter(only_n_count)
only_n_count.most_common(100)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(WordCloud(background_color='white').generate_from_frequencies(only_n_count))
plt.axis('off')
plt.show()
```
